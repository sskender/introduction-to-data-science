```python
# load data from cnt for person 1
raw_data_normal = mne.io.read_raw_cnt("data/1/Normal state.cnt", preload=True, verbose=True).load_data()
raw_data_fatigue = mne.io.read_raw_cnt("data/1/Fatigue state.cnt", preload=True, verbose=True).load_data()
```

```python
# display raw data for normal state
raw_data_normal
```

```python
# display raw data for fatigue state
raw_data_fatigue
```

```python
# display general info and channels
print(raw_data_normal.info)
print(raw_data_normal.info["ch_names"])

print(raw_data_fatigue.info)
print(raw_data_fatigue.info["ch_names"])
```

```python
raw_data_normal.describe()
```

```python
raw_data_fatigue.describe()
```

```python
# extract good channels
goods_ch = set(["T6", "P3", "TP7", "O1", "Oz", "T4", "T5", "FCz", "FC3", "CP3"])
all_ch = set(raw_data_normal.info["ch_names"])
bads_ch = list(all_ch - goods_ch)

raw_data_normal.info["bads"].extend(bads_ch)
raw_data_fatigue.info["bads"].extend(bads_ch)

print(raw_data_normal.info["bads"])
print(raw_data_fatigue.info["bads"])
```

```python
#raw_data_normal.plot()
```

```python
# montage normal state
montage = mne.channels.make_standard_montage("standard_1020")

raw_data_normal.pick_channels(goods_ch)
raw_data_normal.set_montage(montage, match_case=False)
raw_data_normal.plot_sensors(ch_type="eeg")
```

```python
# montage fatigue state
raw_data_fatigue.pick_channels(goods_ch)
raw_data_fatigue.set_montage(montage, match_case=False)
raw_data_fatigue.plot_sensors(ch_type="eeg")
```
